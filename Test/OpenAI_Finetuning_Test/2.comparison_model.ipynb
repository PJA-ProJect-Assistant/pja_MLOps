```python
import requests
import time

base_url = "http://3.34.185.3:8000/"

def ERD_generate(user_info, requirement, project_info, base_url = base_url) :

  start_time = time.time()

  data = {
    "project_overview": user_info,
    "requirements" : requirement,
    "project_summury": project_info,
  }
  try :
    response = requests.post(
      url = base_url + "api/PJA/json_ERD/generate",
      json = data,
      headers= {"Content-Type": "application/json"}
    )

    end_time = time.time()
    response_time = start_time - end_time
    print(response_time)

    return response.json()['json']
  
  except:
    return None


def new_ERD_generate(user_info, requirement, project_info, base_url = base_url) :

  data = {
    "project_overview": user_info,
    "requirements" : requirement,
    "project_summury": project_info,
    "model": "ft:gpt-4o-mini-2024-07-18:test:pja-api-finetuning-model:BmOdcDUE"
  }

  start_time = time.time() 

  try :
    response = requests.post(
      url = base_url + "api/PJA/json_API/generate",
      json = data,
      headers= {"Content-Type": "application/json"}
    )

    end_time = time.time()
    response_time = start_time - end_time
    print(response_time)

    return response.json()['json']

  except:
    return None
```
