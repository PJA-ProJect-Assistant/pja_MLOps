```python
df = df_138[['user_input', 'total_requirements']]
df['total_requirements'] = df['total_requirements'].str.replace(r'(FUNCTIONAL|PERFORMANCE)\d+', r'\1', regex=True)
df.head()
```

1. LLM으로 데이터 생성
- 어처피 모델 이제 안고칠꺼 같으니까 지금모델 확인겸 돌리기
- 로컬로 FastAPI 열고 진행
  - 전체 요약
  - ERD
  - API 명세서 순으로

```python
import requests

base_url = "http://127.0.0.1:8000/"

def project_info_generate(user_info, requirement, base_url = base_url) :
  data = {
    "project_overview": user_info,
    "requirements": requirement,
    "max_tokens": 4000,
    "temperature": 0.3,
    "model": "gpt-4o-mini"
  }
  try :
    response = requests.post(
      url = base_url + "api/PJA/json_Summury/generate",
      json = data,
      headers= {"Content-Type": "application/json"}
    )

    return response.json()['json']

  except :
    return None


def ERD_generate(user_info, requirement, project_info, base_url = base_url) :
 
  data = {
    "project_overview": user_info,
    "requirements" : requirement,
    "project_summury": project_info,
    "max_tokens": 4000,
    "temperature": 0.3,
    "model": "gpt-4o-mini"
  }
  try :
    response = requests.post(
      url = base_url + "api/PJA/json_ERD/generate",
      json = data,
      headers= {"Content-Type": "application/json"}
    )

    return response.json()['json']
  
  except :
    return None


def API_generate(user_info, requirement, project_info, base_url = base_url) :

  data = {
    "project_overview": user_info,
    "requirements" : requirement,
    "project_summury": project_info,
    "max_tokens": 4000,
    "temperature": 0.3,
    "model": "gpt-4o-mini"
  }
  try :
    response = requests.post(
      url = base_url + "api/PJA/json_API/generate",
      json = data,
      headers= {"Content-Type": "application/json"}
    )
    return response.json()['json']

  except:
    return None
```

```python
user_info = df['user_input'][0]
requirement = df['total_requirements'][0]
```

```python
project_info_lst = []
ERD_generate_lst = []
API_generate_lst = []

for i in range(len(df)):

  try :
    project_info = project_info_generate(df['user_input'][i], df['total_requirements'][i])
    project_info_lst.append(project_info)

    ERD_generate_lst.append(ERD_generate(df['user_input'][i], df['total_requirements'][i], str(project_info)))
    API_generate_lst.append(API_generate(df['user_input'][i], df['total_requirements'][i], str(project_info)))

  except :
    None

df['project_info'] = project_info_lst
df['ERD_data'] = ERD_generate_lst
df['API_data'] = API_generate_lst
```
